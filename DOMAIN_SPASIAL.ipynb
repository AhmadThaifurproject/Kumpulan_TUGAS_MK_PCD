```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

# Load gambar
img = data.camera()

# Tampilkan gambar asli
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')

# Ambil dimensi gambar
height, width = img.shape

# Operasi dalam domain spasial: Flip Vertikal
img_flipped_vertical = np.flipud(img)

# Tampilkan gambar setelah dilakukan operasi flip vertikal
plt.subplot(122)
plt.imshow(img_flipped_vertical, cmap='gray')
plt.title('Hasil Flip Vertikal')

plt.show()
```
