```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
from skimage import data
from PIL import Image

# Load gambar
img = data.camera()
img_array = np.array(img)

# Transformasi Fourier 2D
frequencies = fftpack.fft2(img_array)

# Pusatkan frekuensi rendah di tengah
frequencies_shifted = fftpack.fftshift(frequencies)

# Ambil magnitudo spektrum frekuensi
magnitude_spectrum = 20 * np.log(np.abs(frequencies_shifted))

# Tampilkan gambar asli
plt.subplot(121)
plt.imshow(img_array, cmap='gray')
plt.title('Gambar Asli')

# Tampilkan spektrum frekuensi
plt.subplot(122)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Spektrum Frekuensi')
plt.colorbar()

plt.show()
```
